```python
import numpy as np
def check_constraint(x, capacity, weight):
    chosen_bin = compute_fitness(x)
    if chosen_bin == 0:
        return False   
    remaining = [capacity] * chosen_bin    
    if remaining[0] < weight[0]:
        return False
    
    remaining[0] -= weight[0]
    i = 0
    for j in range(1, len(weight)):
        
        if (remaining[i] - weight[j] >= 0):
            remaining[i] -= weight[j]
            i = 0
        else:
            i +=1
            if i == chosen_bin:
                return False
            j -= 1
    return True
def compute_fitness(bin):
    return np.sum(bin)

def compute_val_fitness(bin):
    if (check_constraint(real2bin(bin),c, weight) == True):
        return compute_fitness(bin)
    else:
        return 0
    

def real2bin(x):
    sigmoid_values = 1 / (1 + np.exp(-x))
    binary_values = np.where(sigmoid_values >= 0.5, 1, 0)
    return binary_values

def normalize_array(arr):
    norm = np.amax(arr)
    normalized_arr = arr / norm
    return normalized_arr
# Driver program
weight = [42, 69, 67, 57, 93, 90, 38, 36, 45, 42, 33, 79, 27, 57, 44, 84, 86, 92, 46, 38, 85,
33, 82, 73, 49, 70, 59, 23, 57, 72, 74, 69, 33, 42, 28, 46, 30, 64, 29, 74, 41, 49, 55, 98, 80, 32, 25, 38,
82, 30, 35, 39, 57, 84, 62, 50, 55, 27, 30, 36, 20, 78, 47, 26, 45, 41, 58, 98, 91, 96, 73, 84, 37, 93, 91,
43, 73, 85, 81, 79, 71, 80, 76, 83, 41, 78, 70, 23, 42, 87, 43, 84, 60, 55, 49, 78, 73, 62, 36, 44, 94, 69,
32, 96, 70, 84, 58, 78, 25, 80, 58, 66, 83, 24, 98, 60, 42, 43, 43, 39, 97, 57, 81, 62, 75, 81, 23, 43, 50,
38, 60, 58, 70, 88, 36, 90, 37, 45, 45, 39, 44, 53, 70, 24, 82, 81, 47, 97, 35, 65, 74, 68, 49, 55, 52, 94,
95, 29, 99, 20, 22, 25, 49, 46, 98, 59, 98, 60, 23, 72, 33, 98, 80, 95, 78, 57, 67, 53, 47, 53, 36, 38, 92,
30, 80, 32, 97, 39, 80, 72, 55, 41, 60, 67, 53, 65, 95, 20, 66, 78, 98, 47, 100, 85, 53, 53, 67, 27, 22, 61,
43, 52, 76, 64, 61, 29, 30, 46, 79, 66, 27, 79, 98, 90, 22, 75, 57, 67, 36, 70, 99, 48, 43, 45, 71, 100, 88,
48, 27, 39, 38, 100, 60, 42, 20, 69, 24, 23, 92, 32 ];
c = 150
nVars = len(weight)
```

## Search in space eq
$P_{new}=P_{i}+y(i)(P_{i}-P_{i+1})+x(i)(P_{i}-P_{mean})$

```python
#search stage
def create_x_y_x1_y1__():
    #Using numpy vector for faster computational time
    ## Eq. 2
    phi = a_factor * np.pi * np.random.uniform(0, 1, nVars)
    r = phi + R_factor * np.random.uniform(0, 1, nVars)
    xr, yr = r * np.sin(phi), r * np.cos(phi)

    ## Eq. 3
    r1 = phi1 = a_factor * np.pi * np.random.uniform(0, 1, nVars)
    xr1, yr1 = r1 * np.sinh(phi1), r1 * np.cosh(phi1)

    x_list = xr / np.max(xr)
    y_list = yr / np.max(yr)
    x1_list = xr1 / np.max(xr1)
    y1_list = yr1 / np.max(yr1)
    return x_list, y_list, x1_list, y1_list
```

```python
P_new = np.zeros(nVars)
for i in range(0,pop_size,1):
    x_list, y_list, x1_list, y1_list = create_x_y_x1_y1__() #generamos valores aleatorios por el tamaño de pop size
    for j in range(0,len(P_best),1):
        if (j!=len(P_best)-1):
            P_new[j]=P_best[j]+y_list[j]*(P_best[j]-P_best[j+1])+x_list[j]*(P_best[j]-np.mean(P_best))
        else:
            P_new[j]=P_best[j]+y_list[j]*(P_best[j]-P_best[-1])+x_list[j]*(P_best[j]-np.mean(P_best))
    P_new=normalize_array(P_new)
    if (check_constraint(real2bin(P_new),c, weight) == True):
        if(compute_fitness(real2bin(P_new))<compute_fitness(real2bin(P_best))):
            P_best=P_new
print(check_constraint(real2bin(P_best), c, weight))
print(compute_fitness(real2bin(P_best)))
print(P_best)
print(real2bin(P_best))
```
